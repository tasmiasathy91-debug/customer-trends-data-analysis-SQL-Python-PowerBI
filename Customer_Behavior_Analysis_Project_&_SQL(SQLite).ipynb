# Option 1: Using pd.cut() (Recommended)

```python
#create a column age_group
import pandas as pd

# Create age groups
df['age_group'] = pd.cut(df['age'],
                         bins=[0, 18, 30, 45, 60, 100],
                         labels=['Teen', 'Young Adult', 'Adult', 'Mid Age', 'Senior'])

df[['age', 'age_group']].head(10)
```

# Option 2: Using Custom Logic (apply)

```python
def categorize_age(age):
    if age <= 18:
        return 'Teen'
    elif age <= 30:
        return 'Young Adult'
    elif age <= 45:
        return 'Adult'
    elif age <= 60:
        return 'Mid Age'
    else:
        return 'Senior'

df['age_group'] = df['age'].apply(categorize_age)
```

# Create SQL Environment (SQLite)

```python
import sqlite3

# Create in-memory database
conn = sqlite3.connect(':memory:')

# Load dataframe into SQL table
df.to_sql('customers', conn, index=False, if_exists='replace')
```

# What is the total revenue generated by male vs. female customers

```python
query = """
SELECT Gender,
       SUM(`purchase_amount_(usd)`) AS total_revenue
FROM customers
GROUP BY Gender;
"""

pd.read_sql(query, conn)
```

# Which customers used a discount but still spent more than the average purchase amount

```python
query = """
SELECT *
FROM customers
WHERE discount_applied = 'Yes'
AND `purchase_amount_(usd)` > (
    SELECT AVG(`purchase_amount_(usd)`) FROM customers
);
"""

pd.read_sql(query, conn)
```

What is the top 5 products with the highest average review rating

```python
query = """
SELECT item_purchased,
       AVG(review_rating) AS avg_rating
FROM customers
GROUP BY item_purchased
ORDER BY avg_rating DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

# compare the average purchase amounts between standard and express shipping

```python
query = """
SELECT shipping_type,
       AVG("purchase_amount_(usd)") AS avg_purchase
FROM customers
GROUP BY shipping_type;
"""

pd.read_sql(query, conn)
```

Do subscribed customers spend more ?compare average spend and total revenue between subscribers and non subscribers

```python
query = """
SELECT subscription_status,
       AVG("purchase_amount_(usd)") AS avg_spend,
       SUM("purchase_amount_(usd)") AS total_revenue,
       COUNT(*) AS total_orders
FROM customers
GROUP BY subscription_status;
"""

pd.read_sql(query, conn)
```

# Which 5 products have the highest percentages with discounts applied

```python
query = """
SELECT item_purchased,
       COUNT(*) AS total_orders,
       SUM(CASE WHEN discount_applied = 'Yes' THEN 1 ELSE 0 END) AS discount_count,
       (SUM(CASE WHEN discount_applied = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS discount_percentage
FROM customers
GROUP BY item_purchased
ORDER BY discount_percentage DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

# segment customers into new,returning and loyal based on their total number of previous purchases and show the count of each segment

```python
query = """
SELECT
    CASE
        WHEN previous_purchases = 1 THEN 'New'
        WHEN previous_purchases BETWEEN 1 AND 10 THEN 'Returning'
        ELSE 'Loyal'
    END AS customer_segment,
    COUNT(*) AS customer_count
FROM customers
GROUP BY customer_segment;
"""

pd.read_sql(query, conn)
```

# what are the top 3 most purchased products within each category

```python
query = """
SELECT category, item_purchased, purchase_count
FROM (
    SELECT category,
           item_purchased,
           COUNT(*) AS purchase_count,
           ROW_NUMBER() OVER (PARTITION BY category ORDER BY COUNT(*) DESC) AS rank
    FROM customers
    GROUP BY category, item_purchased
)
WHERE rank <= 3;
"""

pd.read_sql(query, conn)
```

# 9. Are customers who are repeat buyers (more than 5 previous purchases) also likely to subscribe?

```python
query = """
SELECT
    CASE
        WHEN previous_purchases > 5 THEN 'Repeat Buyer'
        ELSE 'Non-Repeat Buyer'
    END AS buyer_type,
    subscription_status,
    COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (PARTITION BY
        CASE
            WHEN previous_purchases > 5 THEN 'Repeat Buyer'
            ELSE 'Non-Repeat Buyer'
        END
    ) AS percentage
FROM customers
GROUP BY buyer_type, subscription_status;
"""

pd.read_sql(query, conn)
```

10.what is the revenue contribution of each age group

```python
query = """
SELECT age_group,
       SUM("purchase_amount_(usd)") AS total_revenue,
       SUM("purchase_amount_(usd)") * 100.0 /
       (SELECT SUM("purchase_amount_(usd)") FROM customers) AS revenue_percentage
FROM customers
GROUP BY age_group;
"""

pd.read_sql(query, conn)
```
